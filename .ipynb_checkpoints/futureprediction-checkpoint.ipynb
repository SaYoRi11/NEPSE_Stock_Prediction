# Predicting the NTC stock prices of the next 60 days 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pickle import load
```

```python
dataset = pd.read_csv('./data/stockdata_test.csv', index_col=0)
inputs = dataset.iloc[len(dataset)-60:,2:3].values
```

```python
sc = load(open('scaler.pkl','rb'))
```

```python
from keras.models import load_model
model = load_model('stockpredictor')
```

```python
for i in range(60):
    inputs = inputs.reshape(-1,1)
    scaled_inputs = sc.transform(inputs)
    X=[]
    X.append(scaled_inputs[i:,0])
    X = np.array(X)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    pred = model.predict(X)
    pred = sc.inverse_transform(pred)
    inputs = np.append(inputs, [pred[0]])
```

```python
future_prediction = inputs[60:]
```

```python
df = pd.DataFrame(future_prediction,columns=['Open'])
```
